#1. Import libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

#2. Load dataset

*   Load dataset (named ***student-dataset.csv***) using ***read_csv*** from pandas
*   Then, display 10 last examples


```python
# Lấy ra dữ liệu từ file trong GG drive đã cấp đường dẫn
dataset = pd.read_csv("student-dataset.csv")

# In ra 10 dữ liệu ở cuối cùng
dataset.tail(10)
```

#3. Show statistics of the given dataset

```python
# Hiển thị số liệu thống kê của tập số liệu tại 2
dataset.describe()
```

#4. Sort dataset by ***nationality***

```python
# Lọc tập dữ liệu theo "nationality"
dataset.sort_values("nationality")
```

#5. Group dataset by ***nationality*** and ***gender***

```python
dataset.groupby(["nationality","gender"]).groups
```

#6. Use scatter plot to represent the relationship between ***age*** and ***refletter.rating***
Rememer adding titles, xlabel, ylabel, ... to the plot

```python
# Tạo ra 1 biểu đồ biểu diễn MQH giữa 'age' và 'refletter.rating'
# Bỏ marker= "*" thì sẽ mặc định là tròn
# s là kích thước của marker
plt.scatter(dataset['age'], dataset['refletter.rating'], label="relationship", color="black", marker="*", s=40)

plt.xlabel('x - age')
plt.ylabel('y - refletter.rating')
plt.title('Relationship between age and refletter.rating')
plt.legend()

plt.show()
```

#7. Use scatter plot to represent the relationship between ***age*** and ***sciences.grade***
Rememer adding titles, xlabel, ylabel, ... to the plot

```python
plt.scatter(dataset['age'], dataset['sciences.grade'], label= "relationship", color="black", marker="H", s = 30)

plt.xlabel('x - age')
plt.ylabel('y - sciences.grade')
plt.title('Relationship between age and sciences.grade')
plt.legend()

plt.show()
```

#8. Use histogram plot to represent the distribution of ***Age***
using bins=5

```python
# 'bin' là số lượng khoảng mà bạn muốn dữ liệu được chia thành trên trục x
# của biểu đồ histogram.
bins = 5

# 'rwidth' xác định chiều rộng của mỗi cột trong biểu đồ histogram, được tính bằng phần trăm chiều rộng của mỗi bin.
plt.hist(dataset['age'], bins, color= 'black', histtype= 'bar', rwidth=0.8)

plt.xlabel('x - age')
plt.ylabel('y - number of human')
plt.title('The distribution of Age')
plt.show()
```

#9. Create a data frame that computes the average of ages in each countries?

Hint: Use groupby, select the age column and aggregate using mean.

```python
# Quoc gia          Trung binh
# Viet name             25
# Nhat                  27
dataset.groupby('nationality')['age'].mean()
```

#10. Create a dataframe including average of ***english.grade, math.grade, sciences.grade, language.grade, portfolio.rating, coverletter.rating, refletter.rating*** of each country


```python
dataset.groupby('nationality')['english.grade','math.grade','sciences.grade','language.grade','portfolio.rating','coverletter.rating','refletter.rating'].mean()
```
