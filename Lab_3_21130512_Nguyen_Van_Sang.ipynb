# Task 2.1. dataset1.csv


```python
# Bước 1: Đọc dữ liệu từ tập tin dataset1.csv
dataset1 = pd.read_csv("dataset1.csv")

# Bước 2: Tách các đặc trưng (X) và biến mục tiêu (y)
X = dataset1[['x']]
y = dataset1['y']

# Bước 3: Huấn luyện mô hình Hồi quy Tuyến tính
regressor = LinearRegression()
regressor.fit(X, y)

# Bước 4: Dự đoán biến mục tiêu bằng cách sử dụng mô hình đã huấn luyện
y_pred = regressor.predict(X)

# Bước 5: Trực quan hóa dữ liệu và đường hồi quy
plt.scatter(X, y, color='blue')
plt.plot(X, y_pred, color='red')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Hồi quy tuyến tính trên Dataset 1')
plt.show()
```

```python

df = pd.DataFrame({'y_pred': y_pred, 'y': y})
print(df)
```

# Task 2.2. dataset2.csv


```python
# Bước 1: Đọc dữ liệu từ tập tin dataset2.csv
dataset2 = pd.read_csv("dataset2.csv")

# Bước 2: Tách các đặc trưng (X) và biến mục tiêu (y)
X = dataset2[['x1', 'x2', 'x3']]
y = dataset2['y']

# Bước 3: Chia tập dữ liệu thành tập huấn luyện và tập kiểm tra
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Bước 4: Huấn luyện mô hình Hồi quy Tuyến tính trên tập huấn luyện
regressor = LinearRegression()
regressor.fit(X_train, y_train)

# Bước 5: Dự đoán biến mục tiêu trên tập kiểm tra
y_pred = regressor.predict(X_test)

# Bước 6: Đánh giá mô hình
print('Explained Variance Score:', metrics.explained_variance_score(y_test, y_pred))
print('Mean Absolute Error:', metrics.mean_absolute_error(y_test, y_pred))
print('Mean Squared Error:', metrics.mean_squared_error(y_test, y_pred))
print('Median Absolute Error:', metrics.median_absolute_error(y_test, y_pred))
print('R^2 Score:', metrics.r2_score(y_test, y_pred))

# Bước 7: Tạo DataFrame từ các dự đoán 'y_pred' và giá trị thực 'y_test'
df = pd.DataFrame({'y_pred': y_pred, 'y_test': y_test})

# Hiển thị DataFrame
print(df.head())
```

# Task 2.4. Apply LinearRegression to the following dataset:
*   Energy efficiency: https://archive.ics.uci.edu/ml/datasets/Energy+efficiency.




```python
# Bước 1: Đọc dữ liệu từ URL
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/00242/ENB2012_data.xlsx"
du_lieu_hieu_suat_nang_luong = pd.read_excel(url)

# Hiển thị thông tin cơ bản về dữ liệu
print(du_lieu_hieu_suat_nang_luong.info())

# Bước 2: Xem một số dòng đầu của dữ liệu
print(du_lieu_hieu_suat_nang_luong.head())

# Bước 3: Xem thống kê mô tả của các biến số
print(du_lieu_hieu_suat_nang_luong.describe())

# Bước 4: Kiểm tra giá trị thiếu
print(du_lieu_hieu_suat_nang_luong.isnull().sum())

# Bước 5: Trực quan hóa mối quan hệ giữa các biến số và hiệu suất năng lượng
sns.pairplot(du_lieu_hieu_suat_nang_luong, x_vars=du_lieu_hieu_suat_nang_luong.columns[:-2], y_vars=['Y1', 'Y2'], kind='reg')

# Bước 6: Tách các đặc trưng (X) và biến mục tiêu (y)
X = du_lieu_hieu_suat_nang_luong.drop(['Y1', 'Y2'], axis=1)
y = du_lieu_hieu_suat_nang_luong[['Y1', 'Y2']]

# Bước 7: Chia tập dữ liệu thành tập huấn luyện và tập kiểm tra
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Bước 8: Huấn luyện mô hình Linear Regression trên tập huấn luyện
regressor = LinearRegression()
regressor.fit(X_train, y_train)

# Bước 9: Dự đoán biến mục tiêu trên tập kiểm tra
y_pred = regressor.predict(X_test)

# Bước 10: Đánh giá mô hình
print('Explained Variance Score (Y1):', metrics.explained_variance_score(y_test['Y1'], y_pred[:, 0]))
print('Mean Absolute Error (Y1):', metrics.mean_absolute_error(y_test['Y1'], y_pred[:, 0]))
print('Mean Squared Error (Y1):', metrics.mean_squared_error(y_test['Y1'], y_pred[:, 0]))
print('Median Absolute Error (Y1):', metrics.median_absolute_error(y_test['Y1'], y_pred[:, 0]))
print('R^2 Score (Y1):', metrics.r2_score(y_test['Y1'], y_pred[:, 0]))

print('Explained Variance Score (Y2):', metrics.explained_variance_score(y_test['Y2'], y_pred[:, 1]))
print('Mean Absolute Error (Y2):', metrics.mean_absolute_error(y_test['Y2'], y_pred[:, 1]))
print('Mean Squared Error (Y2):', metrics.mean_squared_error(y_test['Y2'], y_pred[:, 1]))
print('Median Absolute Error (Y2):', metrics.median_absolute_error(y_test['Y2'], y_pred[:, 1]))
print('R^2 Score (Y2):', metrics.r2_score(y_test['Y2'], y_pred[:, 1]))

# Bước 11: Trực quan hóa dữ liệu và đường hồi quy
plt.scatter(y_test['Y1'], y_pred[:, 0])
plt.xlabel("True Values (Y1)")
plt.ylabel("Predictions (Y1)")
plt.title("True Values vs Predictions (Y1)")
plt.show()

plt.scatter(y_test['Y2'], y_pred[:, 1])
plt.xlabel("True Values (Y2)")
plt.ylabel("Predictions (Y2)")
plt.title("True Values vs Predictions (Y2)")
plt.show()
```
